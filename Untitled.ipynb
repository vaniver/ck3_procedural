```python
map_rad = 2
num_hexes = 3*(map_rad*(map_rad+1))+1
s3 = 1.7
valid_cubes = []
cube2rgb = {}
for i in range(-map_rad,map_rad+1):
    for j in range(-map_rad,map_rad+1):
        k = 0 - i - j
        if -map_rad <= k <= map_rad:
            valid_cubes.append(Cube(i,j,k))
            cube2rgb[Cube(i,j,k)] = (128 + 32 * i,128 + 32 * j,128 + 32 * k)

def cube_round(i, j, k):
    ri, rj, rk = int(round(i)), int(round(j)), int(round(k))
    di, dj, dk = abs(ri-i), abs(rj-j), abs(rk-k)

    if di > dj and di > dk:
        ri = -rj-rk
    elif dj > dk:
        rj = -ri-rk
    else:
        rk = -ri-rj
    return Cube(ri, rj, rk)

max_x, max_y = 500, 500
mid_x, mid_y = max_x//2, max_y//2
for hex_size in range(20, 60):
    cx = np.zeros(num_hexes)
    cy = np.zeros(num_hexes)
    img_provinces = PIL.Image.new('RGB', (max_x,max_y),  "black")
    pix = img_provinces.load()
    c2cx = {}
    c2cy = {}
    c2num = {}
    for ind, c in enumerate(valid_cubes):
        i = c.x
        j = c.y
        k = c.z
        cx[ind] = int(i * 1.5 * hex_size + mid_x)
        cy[ind] = max_y - int((j * 3 * hex_size + cx[ind] - mid_x) / s3 + mid_y)
        c2cx[c] = cx[ind]
        c2cy[c] = cy[ind]
        c2num[c] = 0
    
    def pixel_to_cube(x, y):
        q = (x-mid_x) * 2.0/3.0/hex_size
        r = (-(x-mid_x) / 3.0 + s3/3.0 * (y-mid_y))/hex_size
        return cube_round(q,-q-r,r)
    
    for x in range(max_x):
        for y in range(max_y):
            c = pixel_to_cube(x,y)
            if c in c2cx:
                pix[x,y] = cube2rgb[c]
                c2num[c] += 1
            else:
                continue
    img_provinces.save(os.path.join(data_dir, "testing_sizes", "provinces_hex_%d.png" % hex_size))
    print(hex_size, len(set(c2num.values())))
```
